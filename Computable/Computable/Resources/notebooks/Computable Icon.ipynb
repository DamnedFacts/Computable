## The Computable icon

Abiding to a peculiar software tradition of recursion and bootstrapping things out of thin air, Computable's icon has been created with Computable itself. The colorful, organic shape is generated by a very simple formula:

$$\frac{z}{cos(1-z^{0.4})}$$

To turn this into a picture, we're using a technique called _domain coloring_ on complex-valued fuctions. The code in this notebook has been borrowed largely from Emilia Petrisor's excellent notebook [Visualizing complex-valued functions with Matplotlib and Mayavi](http://nbviewer.ipython.org/github/empet/Math/blob/master/DomainColoring.ipynb).

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb
```

```python
def Hcomplex(z):# computes the hue corresponding to the complex number z
    H = np.angle(z)/(2*np.pi)+1
    return np.mod(H,1)
```

```python
def perfract(x, t, m, M):
    x = x/t
    return m+(M-m)*(x-np.floor(x))
```

```python
def domaincol_co(w,s):
    H = Hcomplex(w)
    m = 0.7 # brightness is restricted to [0.7,1]; interval suggested by E Wegert
    M = 1
    n = 15 # n=number of isochromatic lines per cycle
    isol = perfract(H, 1.0/n, m, M) # isochromatic lines
    modul = np.absolute(w)
    Logm = np.log(modul)
    Logm = np.nan_to_num(Logm)
    modc = perfract(Logm, 2*np.pi/n, m, M)# lines of constant log-modulus
    
    V = modc*isol
    S = 0.9*np.ones_like(H, float)
    HSV = np.dstack((H,S,V))
    RGB = hsv_to_rgb(HSV)
    
    return RGB
```

```python
def func_vals(f, re, im,  N): #evaluates the complex function at the nodes of the grid
   #re and im are  tuples, re=(a,b) and im=(c,d), defining the rectangular region
   #N is the number of nodes per unit interval 
   l = re[1]-re[0]
   h = im[1]-im[0]
   resL = N*l #horizontal resolution
   resH = N*h#vertical resolution
   x = np.linspace(re[0], re[1],resL)
   y = np.linspace(im[0], im[1], resH)
   x,y = np.meshgrid(x,y)
   z = x+1j*y
   w = f(z)
   return w 
```

```python
def plot_domain(color_func,f,re=[-1,1],im= [-1,1],s=0.9, N=200):
    w = func_vals(f, re, im, N)
    domc = color_func(w, s)
    plt.imshow(domc, origin="lower")
    plt.axis('off')
```

```python
plt.rcParams['figure.figsize'] = 8,8
ab = (-2*pi,2*pi)
cd = (-2*pi,2*pi)
f = lambda z: z/np.cos(1-z**0.4)
plot_domain(domaincol_co,f,re=ab,im=cd,N=100)
```
